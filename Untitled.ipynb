```python
df=pd.read_csv('Master roster.CSV', sep=',')
```

```python
df
```

```python
# Nominatim geocoder for OpenStreetMap data with RateLimiter
geocoder = RateLimiter(Nominatim(user_agent='tutorial').geocode, min_delay_seconds=1)
df['Hometown'] = df['Hometown'].apply(geocoder)

# add latitude and longitude to dataframe
df['Latitude'] = df['Hometown'].apply(lambda loc: loc.latitude if loc else None)
df['Longitude'] = df['Hometown'].apply(lambda loc: loc.longitude if loc else None)

df
```

```python
df.to_csv('Roster Geo.csv')
```
